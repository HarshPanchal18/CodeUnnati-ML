```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y = [2, 4, 1]

plt.plot(x, y)

plt.xlabel("X axis")
plt.ylabel("Y axis")

plt.show()
```

```python
y2019 = [2, 3, 4.5]
y2020 = [1, 1.5, 5]
x = ["iPhone X", "iPhone 11", "iPhone 12"]

plt.plot(x, y2019, c="blue", marker="o")
plt.plot(
    x,
    y2020,
    c="green",
    marker="o",
    linestyle="dashed",
    markersize=4,
    mec="r",
)

plt.legend(["year 2019", "year 2020"], loc="lower right")
plt.xlabel("Apple iPhone models")
plt.ylabel("Total sales")

plt.show()
```

```python
names = ["Sonu", "Raj", "Divya"]
marks = [87, 50, 98]


plt.figure(figsize=(13, 3))
plt.subplot(131)
plt.bar(x=names, height=marks)
plt.title("Plotting Bar Graph")
plt.xlabel("Names")
plt.ylabel("Marks")

plt.subplot(132)
plt.scatter(x=names, y=marks)
plt.title("Plotting Scatter Graph")
plt.xlabel("Names")
plt.ylabel("Marks")

plt.subplot(133)
plt.plot(names, marks)
plt.title("Plotting Line Graph")
plt.xlabel("Marks")
plt.ylabel("Marks")

plt.suptitle("Categorical Plotting")

plt.show()
```
